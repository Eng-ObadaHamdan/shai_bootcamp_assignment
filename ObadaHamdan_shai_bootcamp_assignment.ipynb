# 1- **Basic** **Data** **Exploration**: Identify the number of rows and columns in the dataset, determine the data types of each column, and check for missing values in each column.

# **2-Descriptive Statistics**: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.

```python
import pandas as pd
import numpy as np


mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

salary_range = max_salary - min_salary

std_dev_salary = df['TotalPay'].std()

print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
print("Minimum Salary:", min_salary)
print("Maximum Salary:", max_salary)
print("Salary Range:", salary_range)
print("Standard Deviation of Salary:", std_dev_salary)
```

# **3- Data Cleaning:** Handle missing data by suitable method with explain why you use it
# I replaced the missing data with the column's arithmetic mean, because it gives me the average of the data and the value will often be close to this number.

```python
import pandas as pd

df = pd.read_csv('/content/Salaries.csv')



# Compute the mean of the 'BasePay' column
mean_basepay = df.loc[df['BasePay'] > 0, 'BasePay'].mean()

# Replace values less than or equal to zero with the mean
df.loc[df['BasePay'] <= 0, 'BasePay'] = mean_basepay
# Replace null values with the mean
df['BasePay'].fillna(mean_basepay, inplace=True)



# Compute the mean of the 'OvertimePay' column
mean_OvertimePay = df.loc[df['OvertimePay'] > 0, 'OvertimePay'].mean()

# Replace values less than to  zero with the mean
df.loc[df['OvertimePay'] < 0, 'OvertimePay'] = mean_OvertimePay
# Replace null values with the mean
df['OvertimePay'].fillna(mean_OvertimePay, inplace=True)



# Compute the mean of the 'OtherPay' column
mean_OtherPay = df.loc[df['OtherPay'] > 0, 'OtherPay'].mean()

# Replace values less than  to zero with the mean
df.loc[df['OtherPay'] < 0, 'OtherPay'] = mean_OtherPay
# Replace null values with the mean
df['OtherPay'].fillna(mean_OtherPay, inplace=True)



# Compute the mean of the 'Benefits' column
mean_Benefits = df.loc[df['Benefits'] > 0, 'Benefits'].mean()

# Replace values less than  to zero with the mean
df.loc[df['Benefits'] < 0, 'Benefits'] = mean_Benefits
# Replace null values with the mean
df['OtherPay'].fillna(mean_Benefits, inplace=True)


# Compute the mean of the 'TotalPay' column
mean_TotalPay = df.loc[df['TotalPay'] > 0, 'TotalPay'].mean()

# Replace values less than or equal to zero with the mean
df.loc[df['TotalPay'] <= 0, 'TotalPay'] = mean_TotalPay
# Replace null values with the mean
df['TotalPay'].fillna(mean_TotalPay, inplace=True)



# Compute the mean of the 'TotalPayBenefits' column
mean_TotalPayBenefits = df.loc[df['TotalPayBenefits'] > 0, 'TotalPayBenefits'].mean()

# Replace values less than or equal to zero with the mean
df.loc[df['TotalPayBenefits'] <= 0, 'TotalPayBenefits'] = mean_TotalPayBenefits
# Replace null values with the mean
df['TotalPayBenefits'].fillna(mean_TotalPayBenefits, inplace=True)





```

# **4-Basic Data Visualization:** Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
import matplotlib.pyplot as plt

# Create a histogram to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Create a pie chart to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(10)  # Considering top 10 departments for better visualization

plt.figure(figsize=(10, 8))
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Different Departments')
plt.axis('equal')
plt.show()
```

# 5-Grouped Analysis: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python
import pandas as pd
import matplotlib.pyplot as plt  # Add this line to import matplotlib

df = pd.read_csv('/content/Salaries.csv')

# Group the data by 'JobTitle' and calculate summary statistics for each group
grouped_stats = df.groupby('JobTitle')['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])

# Sort the groups by average salary in descending order
grouped_stats = grouped_stats.sort_values(by='mean', ascending=False)

# Print the summary statistics for each group
print("Summary Statistics for Each Job Title:")
print(grouped_stats)

# Plot average salary across different groups (in this case, job titles)
plt.figure(figsize=(12, 8))
grouped_stats['mean'].plot(kind='bar', color='skyblue')
plt.title('Average Salary Across Different Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

## **6- Simple Correlation Analysis:** Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Salaries.csv')


# Calculate the correlation coefficient
correlation = df['TotalPay'].corr(df['OvertimePay'])
print("Correlation between TotalPay and OvertimePay:", correlation)

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['OvertimePay'], df['TotalPay'], alpha=0.5)
plt.title('Scatter Plot: TotalPay vs OvertimePay')
plt.xlabel('OvertimePay')
plt.ylabel('TotalPay')
plt.grid(True)
plt.show()
```
